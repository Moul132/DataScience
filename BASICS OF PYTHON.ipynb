# Exercise 1: Prime Numbers

````python
def is_prime(num): #Creating Function
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

number = int(input("Enter a number to check if it's prime: ")) #User Input
if is_prime(number): # Callinig function
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
````

# Exercise 2: Product of Random Numbers

````python
import random #Importing random to generate random numbers by definig ranges of numbers we need

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)
correct_product = num1 * num2

user_answer = int(input(f"Enter the product of {num1} and {num2}: "))
if user_answer == correct_product: #Checking user answer with correct product
    print("Correct Answer")
else:
    print(f"Incorrect. The correct answer is {correct_product}.")
````

#Exercise 4: Word Counter

````python
from collections import Counter
import string

# Sample input text
input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Remove punctuation and convert given text to lowercase
cleaned_text = input_text.translate(str.maketrans('', '', string.punctuation)).lower()
print(cleaned_text)
# Split text into wo
words = cleaned_text.split()
print("\nwords")

word_counts = Counter(words) # Count occurrences of each word

for word, count in word_counts.items(): # Print word counts
    print(f"'{word}': {count}")
````

# Exercise 5: Check for Palindrome

````python
input_string = input("Enter a string to check if it's a palindrome: ")
if input_string == input_string[::-1]:
    print(f"'{input_string}' is a palindrome.")
else:
    print(f"'{input_string}' is not a palindrome.")
````
