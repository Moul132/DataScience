# **Estimation And Confidence Intervals**

**GIVEN DATA**

```
# A total of 15 print-heads were randomly selected and tested until failure.The durability of each print-head (in millions of characters)was recorded as follows:

1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29

**TASKS**
a. Build 99% Confidence Interval Using Sample Standard Deviation

b.Build 99% Confidence Interval Using Known Population Standard Deviation(Population Std=0.2)
```



# a. Build 99% Confidence Interval Using Sample Standard Deviation

````python
import numpy as np
from scipy.stats import t

# Given data
Records= [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
n=len(Records)
print('sample_size: ' , n)

# Calculate sample mean and standard deviation
sample_mean = np.mean(Records)
sample_std_dev = np.std(Records, ddof=1)  # ddof=1 for sample standard deviation
print('Sample_mean: ' ,sample_mean)
print('Sample_std_dev: ' ,sample_std_dev)

# Determine degrees of freedom
degrees_of_freedom = n - 1

# Calculate critical t-value for 99% confidence interval
alpha = 0.01 # (100-99)/100
t_critical = t.ppf(1 - alpha/2, degrees_of_freedom)
print('t_critical: ' ,t_critical)

# Calculate margin of error
margin_of_error = t_critical * (sample_std_dev / np.sqrt(n))
print('margin_of_error: ' ,margin_of_error)


# Construct 99% confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("99% Confidence Interval:", confidence_interval)
````

# **b.Build 99% Confidence Interval Using Known Population Standard Deviation(Population Std=0.2)**

````python
import numpy as np
from scipy.stats import norm

# Given data
Records = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]

# Determine sample size
n = len(Records)
print('Sample_size: ' ,n)

# Calculate sample mean
sample_mean = np.mean(Records)
print('Sample_mean: ' ,sample_mean)

# Known population standard deviation
population_std_dev=0.2 #Given
print('Population_std_dev: ' ,population_std_dev)

# Calculate critical z-value for 99% confidence interval
alpha = 0.01
z_critical = norm.ppf(1 - alpha/2)
print('z_critical: ' ,z_critical)

# Calculate margin of error
margin_of_error = z_critical * (population_std_dev / np.sqrt(n))
print('margin_of_error: ' ,margin_of_error)

# Construct 99% confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("99% Confidence Interval:", confidence_interval)
````
